## 14. total percentage of active hour for a month 

```python
df=df=pd.read_excel('Agent_Login_Report (4).xls',skiprows=2)
del df['SL No']
df
```

```python
df['Date']=df['Date'].apply(lambda x: pd.to_datetime(x))
```

```python
df['New Duration'] = (pd.to_datetime(df.Date) + pd.to_timedelta(df.Duration))
df.dtypes
df
```

```python
df['Hour']=df['New Duration'].dt.hour
df['Minute']=df['New Duration'].dt.minute
df['Sec']=df['New Duration'].dt.second
```

```python
df['HtoS']=df['Hour']*3600
df['MtoS']=df['Minute']*60
df['Total Seconds']=df['HtoS']+df['MtoS']+df['Sec']
```

```python
df['Total Hours']=df['Total Seconds']/3600
df
```

```python
#Assuming 22 working days in a month. Therefore, total hours=22*24=528 hours

df4=((df.groupby('Agent')['Total Hours'].sum())/528)*100
df5=df4.to_frame('% of Active Hours')
df5
```

```python
for value in df5['% of Active Hours'].values:
    value2=round(value,2)
    value3=str(value2)+'%'
    df5['% of Active Hours']=df5['% of Active Hours'].replace(value,value3)
```
